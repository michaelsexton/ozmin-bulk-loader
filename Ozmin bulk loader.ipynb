```python
import pandas as pd
import cx_Oracle
from ozmin_loader import db
```

```python
connection = db.OzminConnection()
```

```python
nickel_resources = pd.read_excel('nickel/Ni Reserves and Resources 2017.xlsx')
nickel_resources = nickel_resources[nickel_resources["DEPOSIT_NAME"].notnull()]
nickel_resources = nickel_resources[nickel_resources["Date"].notnull()]
nickel_resources = nickel_resources[nickel_resources["PVR_MT"]!="NO CHANGE"]
nickel_resources = nickel_resources[nickel_resources["Update status"]!="OZMIN updated"]
nickel_resources
```

```python
for index, row in nickel_resources.iterrows():
    deposit_eno = row["DEPOSIT_ENO"]
    deposit_name = row["DEPOSIT_NAME"]
    zone_eno = row["ZONE_ENO"]
    zone_name = row["ZONE_NAME"]
    record_date = row["Date"]
    pvr, pbr, mrs, idr, ifr = row[["PVR_MT","PBR_MT","MRS_MT","IDR_MT","IFR_MT"]]

    deposit_eno = connection.get_or_create_deposit(deposit_eno,deposit_name)
    zone_eno = connection.get_or_create_zone(deposit_eno, zone_eno, zone_name)
    sql = connection.add_resources(zone_eno, record_date, pvr, pbr, mrs, idr, ifr, "")
```
